```python
def run_expirement(q=None, Ks= [20,50, 100]):
    bar_width = 0.2
    plt.figure(figsize=(10, 6))
    plt.xlabel('Support')
    plt.ylabel('Weights')
    
    for i, k in enumerate(Ks):
        samples = rng.choice(support, p=pdf, size=k)
        bar_offset = -bar_width*i+0.5*len(Ks)*bar_width
        plt.bar(np.linspace(0,15,16) + bar_offset, sir(support, samples, q=q), width=bar_width, align="center", label=f"{k}")
    plt.plot(np.linspace(0,15,1000), p(np.linspace(0,15,1000)), label = "actual")
    plt.legend()

run_expirement(q=None, Ks= [20,50,100])
run_expirement(q= lambda x : norm.pdf(x, 5, 4), Ks= [20,50,100])
```
