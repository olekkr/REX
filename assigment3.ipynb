We made a program that would count the pixels of detected markers by measuring the height of the left and right side and taking the average. If given distance from marker and height of marker it would then calculate `f`. Below image was taken using the program and capturing with dimensions `(480,270)`, which measured `marker 4` (1000mm) to have a height of `94px`. For `marker 2` (2000mm), it measured `47px`. The measurements were consistently scaled, and we calculated the focal length `f` to be `648`

**1000mm marker 4**

$x = 94$


$X = 145$


$Z = 1000$


$x * (Z / X) = 94 * (1000 / 145) = 648.276$


**2000mm marker 2**

$x = 47$

$X = 145$

$Z = 2000$

$x * (Z / X) = 47 * (2000 / 145) = 648.276 $
