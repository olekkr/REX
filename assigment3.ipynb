We made a program that would count the pixels of detected markers by measuring the height of the left and right side and taking the average. If given distance from marker and height of marker it would then calculate `f`. Below images were taken using the program and capturing with dimensions `(480,270)`. We measured the actual height of the markers to be `145mm`

| x   | X   | Z        | f           |
| --- | --- | -------- | ----------- |
| 183 | 145 | 500      | 631,0344828 |
| 122 | 145 | 750      | 631,0344828 |
| 91  | 145 | 1000     | 627,5862069 |
| 73  | 145 | 1250     | 629,3103448 |
| 60  | 145 | 1500     | 620,6896552 |
| 52  | 145 | 1750     | 627,5862069 |
| 45  | 145 | 2000     | 620,6896552 |
| 40  | 145 | 2250     | 620,6896552 |
| 36  | 145 | 2500     | 620,6896552 |
|     |     | mean     | 625,4789272 |
|     |     | variance | 22,12973973 |

We will use the mean of 625, as we see it has a downwards trend, the further away from the camera, the object is. We believe this is because we are losing resolution the further away the object is. We therefore believe that to get the best overall estimates, we will use the `f` value most likely to be closest to the actual value, since the variance, when taking the sqrt of it, it will become the standard deviation `sqrt(22) = 4.69`, which we can see fits well that some are 620, and some are 630, which are approx +-5 from the mean.
